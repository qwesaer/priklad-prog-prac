### Закон Архимеда

**Закон Архимеда** — один из основных законов гидростатики и аэростатики, описывающий выталкивающую силу, действующую на тело, погружённое в жидкость или газ.

Ключевые пункты:
1. На тело, погружённое в жидкость (или газ), действует выталкивающая сила.
2. Эта сила равна весу жидкости (или газа) в объёме погружённой части тела.
3. Выталкивающая сила направлена вертикально вверх.

Основная формула:

$$ F_A = \rho \cdot g \cdot V $$

где:
- $F_A$ — сила Архимеда (Н),
- $\rho$ — плотность жидкости или газа (кг/м³),
- $g$ — ускорение свободного падения (м/с²),
- $V$ — объём погружённой части тела (м³).
